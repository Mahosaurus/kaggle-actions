```python
results = []
for student_id in test["student_id"]:
    content = random.gauss(0, 1.04)
    wording = random.gauss(content, 0.4)
    results.append([student_id, round(content, 2), round(wording, 2)])
result = pd.DataFrame(results, columns=["student_id", "content", "wording"])
print(result)
result.to_csv("submission.csv", index=False)
```
